# 구분자 변경

```python
# 장르 이름에는 빈칸 없다
# 만약 빈칸 있으면 _ 붙이는 처리 필요
# 예) k drama|fun --> k drama fun  (X)
#                 --> k_drama fun  (O)
movies[movies['genres'].str.contains(' ')]
```

```python
movies['genres'] = movies['genres'].str.replace('|',' ', regex=True)
movies
```

# 연도 무결측 확인

```python
# 제목에 연도 없는 데이터 검색
# movies[~movies['title'].str.contains('(')]  <-- \( 사용해야함
movies[~movies['title'].str.contains('\(')]
```

# 영화명/연도 분리

```python
movies['title'] = movies['title'].str.strip()
movies
```

```python
movies[['title', 'year']] = movies['title'].str.rsplit(' ', n=1, expand=True) # expand로 하면 리스트가 아닌 데이터프레임으로 처리해줌
movies
```

```python
movies['year'] = movies['year'].apply(lambda x: x[1:-1])
movies
```

```python
# for문을 사용하여 '-'가 있는 경우 처리
for index, row in movies.iterrows():
    if '–' in row['year']:
        split_years = row['year'].split('–')
        movies.at[index, 'year'] = split_years[0]
```

```python
movies['year'] = movies['year'].astype(int)
movies
```

```python
# 전처리 베이스라인 코드대로면 같은 제목을 가진 서로 다른 영화가 중복되어 분리해주는 작업 필요
dup_movie = movies[movies.duplicated('title', keep=False)]
movie = movies[~movies.duplicated('title', keep=False)]

dup_movie['title_order'] = dup_movie.groupby('title').cumcount() + 1
dup_movie['title_order'] = dup_movie['title_order'].astype(str)
dup_movie["title"] = dup_movie["title"] + dup_movie["title_order"]
dup_movie = dup_movie.drop(["title_order"], axis=1)
```

```python
dup_movie
```

# movies 저장

```python
# 데이터 저장
merged = pd.concat([movie, dup_movie], ignore_index=True)
merged.to_csv(path + 'movies_final.csv', index=False)
```
