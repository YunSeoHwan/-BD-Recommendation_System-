# 결측 장르 영화 확인

```python
no_genres = df[df['genres'] == '(no genres listed)']
```

```python
df_links = pd.read_csv("D:/kwangwoon/2023_2학기/빅데이터처리및응용/teamproject/links.csv", usecols=['movieId', 'imdbId'])
```

# 장르 결측치 처리

```python
# 장르가 없는 영화 df와 link 데이터프레임을 병합하여 imdbId를 연결
no_genre_movies = pd.merge(no_genres, df_links, on='movieId')
```

```python
# 인덱스를 movieId로 변경
no_genre_movies.index = no_genre_movies['movieId']
```

```python
# movieId 컬럼은 삭제
no_genre_movies.drop('movieId', axis=1, inplace=True)
```

```python
# imdbId 값을 모두 str로 변환
no_genre_movies['imdbId'] = no_genre_movies['imdbId'].astype(str)
```

```python
# imdbId 전처리
no_genre_movies['imdbId'] = no_genre_movies['imdbId'].apply(lambda x: '0' * (7 - len(x)) + x if len(x) < 7 else x)
```

```python
no_genre_movies.info()
```

```python
no_genre_movies
```

```python
# 크롬 드라이버 설치, 연결
s = Service("D:\chromedriver.exe")
driver = webdriver.Chrome(service=s)

# 브라우저 화면 크기 변경하기
driver.maximize_window()
```

```python
# 각 행에 대해 URL 생성 및 접속
for index, row in no_genre_movies.iterrows():
    imdb_id = row['imdbId']
    url = f"https://www.imdb.com/title/tt{imdb_id}/"

    # 해당 URL에 접속
    driver.get(url)
    time.sleep(1)
    # 웹 페이지에서 필요한 작업 수행 가능
    try:
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        genre_list = soup.select('div.ipc-chip-list__scroller > a.ipc-chip.ipc-chip--on-baseAlt')
        result = '|'.join([genre.text for genre in genre_list])
        no_genre_movies.at[index, 'genres'] = result
        
    except:
        print(f'{index} : failed')
        continue
    

    # 예시로 현재 URL 출력
    current_url = driver.current_url
    print(f"Index: {index}, IMDb URL: {current_url}, genres: {result}")

driver.quit()
```

```python
no_genre_movies
```

```python
# df의 'movieId' 컬럼 값과 동일한 no_genre_movies의 인덱스를 찾아 'genres' 컬럼 값을 수정
df.loc[df['movieId'].isin(no_genre_movies.index), 'genres'] = df['movieId'].map(no_genre_movies['genres'])
```

```python
# csv 파일로 저장
df.to_csv('./movie_genre_added.csv', index=False)
```

# 연도 결측치 처리

```python
df = pd.read_csv('./trash/movie_genre_added.csv')
```

```python
df.head()
```

```python
# 뒷자리가 '(' 로 시작하지 않는 row만 추출
no_years = df[~df['title'].str.contains('\(')]
```

```python
# 장르 전처리와 유사하게 진행
no_year_movies = pd.merge(no_years, df_links, on='movieId')

no_year_movies.index = no_year_movies['movieId']

no_year_movies.drop('movieId', axis=1, inplace=True)

no_year_movies['imdbId'] = no_year_movies['imdbId'].astype(str)

no_year_movies['imdbId'] = no_year_movies['imdbId'].apply(lambda x: '0' * (7 - len(x)) + x if len(x) < 7 else x)

no_year_movies.info()
```

```python
# 연도 결측인 데이터들
no_year_movies
```

```python
# 크롬 드라이버 설치, 연결
s = Service("D:\chromedriver.exe")
driver = webdriver.Chrome(service=s)

# 브라우저 화면 크기 변경하기
driver.maximize_window()
```

```python
# 각 행에 대해 URL 생성 및 접속
for index, row in no_year_movies.iterrows():
    imdb_id = row['imdbId']
    url = f"https://www.imdb.com/title/tt{imdb_id}/"

    # 해당 URL에 접속
    driver.get(url)
    time.sleep(1)
    # 웹 페이지에서 필요한 작업 수행 가능
    try:
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        year = soup.select('div.sc-69e49b85-0.jqlHBQ > ul > li> a.ipc-link.ipc-link--baseAlt.ipc-link--inherit-color')[0].text
        no_year_movies.at[index, 'years'] = year
        
    except:
        print(f'{index} : failed')
        continue
    

    # 예시로 현재 URL 출력
    current_url = driver.current_url
    print(f"Index: {index}, IMDb URL: {current_url}, years: {year}")

driver.quit()
```

```python
# 크롤링 되지 않은 한 값만 직접 입력
no_year_movies.loc[147250, 'years'] = '1980'
```

```python
no_year_movies
```

```python
# for문을 사용하여 '-'가 있는 경우 처리
for index, row in no_year_movies.iterrows():
    if '–' in row['years']:
        split_years = row['years'].split('–')
        no_year_movies.at[index, 'years'] = split_years[0]
```

```python
no_year_movies
```

```python
# no_year_movies의 각 행에 대해 title 수정
for index, row in no_year_movies.iterrows():
    movie_id = index
    new_title = row['title'] + f" ({row['years']})" if pd.notna(row['years']) else row['title']
    
    # df 데이터프레임에서 해당 movieId의 title 수정
    mask = (df['movieId'] == movie_id)
    df.loc[mask, 'title'] = new_title
```
